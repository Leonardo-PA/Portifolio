Como seu primeiro trabalho de desenvolvedor em uma empresa de consultoria de software, você foi designado para antender um cliente que fabrica peças automotivas e criar um protótipo da solução que ele necessita. O cliente relata que tem aumentado o número de peças e que gostaria de uma solução que fosse capaz de prever quantas peças serão vendidas em um determinado mês. Por se tratar de um protótipo, você deve utilizar somente as informações que lhe foram cedidas, não precisa, nesse momento, analisar o comportamento de fatores externos, por exemplo, comportamento da bolsa de valores, tendência de mercado, etc.

apos obter os dados podemos verificar que se trata de uma constante que podemos usar uma função de primeiro grau para prever qual será o resultado nos proximos meeses, a função que podemos usar para calcular é :  R = C * Mes, onde:
R resultado, C contante, mes a variavel de entrada, assim podendo obter qualquer resultado aproximado.


teste no caderno da funcionalidade da equação;

mes = 2 ;  c = 2000 -> r = 200*2 = 400 (Valor para o mes 2 )

mes = 3 ;  c = 2000 -> r = 200*3 = 600 (Valor para o mes 3 )

proximo passo implementar a solução no codigo e aplicar.

```python
c= 200 # valor da constante.

mes = input("digite o mês que deseja saber o  resultado: ") # função para capturar o mes que o cliente digitar.

mes = int(mes) # converter o valor par numerico obtido pela função do input()

r = c * mes # equação de primeiro grau ou função linear.

print(f"A quantidade de peças para o mês {mes} será {r}") # Impressão do resultado usando uma string intekrpolada com "f-string"
```
